```python
import numpy as np

# 3. Масив 5x5 з випадковими числами від -1 до 1
matrix2 = np.random.uniform(-1, 1, (5, 5))
print("Матриця 5x5 [-1,1]:\n", matrix2)
```

```python
import numpy as np

# 4. Масив 3x3 з випадковими числами від 0 до 1
matrix3 = np.random.rand(3, 3)
print("Матриця 3x3 [0,1]:\n", matrix3)
```

```python
import numpy as np

# 5. Два масиви по 6 елементів та поелементні операції
arr2 = np.random.randint(1, 21, 6)
arr3 = np.random.randint(1, 21, 6)

print("Масив 1:", arr2)
print("Масив 2:", arr3)

print("Додавання:", arr2 + arr3)
print("Віднімання:", arr2 - arr3)
print("Множення:", arr2 * arr3)
```

```python
import numpy as np

# 6. Множення матриць 3x2 і 2x4
A = np.random.randint(1, 6, (3, 2))
B = np.random.randint(1, 6, (2, 4))

print("Матриця A:\n", A)
print("Матриця B:\n", B)

result = np.dot(A, B)
print("Результат множення:\n", result)
```

```python
import numpy as np

# 7. Транспонування матриці 3x3
matrix4 = np.random.rand(3, 3)
print("Початкова матриця:\n", matrix4)

transpose = matrix4.T
print("Транспонована матриця:\n", transpose)
```

```python
import numpy as np

# 8. Різниця двох матриць 3x3
matrix5 = np.random.randint(1, 11, (3, 3))
matrix6 = np.random.randint(1, 11, (3, 3))

print("Матриця 1:\n", matrix5)
print("Матриця 2:\n", matrix6)

diff = matrix5 - matrix6
print("Різниця:\n", diff)
```

```python
import numpy as np

# 9. Добуток двох матриць 3x3
matrix7 = np.random.randint(-5, 6, (3, 3))
matrix8 = np.random.randint(-5, 6, (3, 3))

print("Матриця 1:\n", matrix7)
print("Матриця 2:\n", matrix8)

product = np.dot(matrix7, matrix8)
print("Добуток:\n", product)
```

```python
import numpy as np

# 10. Сума елементів кожного рядка (вектор-стовпчик)
matrix9 = np.random.uniform(-1, 1, (3, 3))
print("Матриця:\n", matrix9)

row_sums = np.sum(matrix9, axis=1).reshape(-1, 1)
print("Вектор-стовпчик сум:\n", row_sums)
```
