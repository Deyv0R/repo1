```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_books = pd.read_csv('/content/drive/MyDrive/КСВД/bestsellers with categories.csv', encoding='latin-1')

# Розмір
print("Розмір:", df_books.shape)

# Скільки книг
print("Кількість книг:", len(df_books))

# Зміна назв колонок
df_books.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']

# Пропуски
print(df_books.isna().sum())

print("Чи є пропуски:", "Так" if df_books.isna().sum().sum() > 0 else "Ні")

# Унікальні жанри
print("Жанри:", df_books['genre'].unique())

# Статистика цін
print("Макс:", df_books['price'].max())
print("Мін:", df_books['price'].min())
print("Середня:", df_books['price'].mean())
print("Медіана:", df_books['price'].median())

# Найвищий рейтинг
max_rating = df_books['user_rating'].max()
print("Найвищий рейтинг:", max_rating)

# Скільки книг з таким рейтингом
print("Кількість:", len(df_books[df_books['user_rating'] == max_rating]))

# Книга з найбільшою кількістю відгуків
top_reviews = df_books.loc[df_books['reviews'].idxmax()]
print("Найбільше відгуків:", top_reviews['name'])

# Найдорожча книга 2010 року
books_2010 = df_books[df_books['year'] == 2010]
print("Найдорожча 2010:", books_2010.loc[books_2010['price'].idxmax()]['name'])

# Скільки Fiction у 2012
fiction_2012 = df_books[(df_books['year'] == 2012) & (df_books['genre'] == 'Fiction')]
print("Fiction 2012:", len(fiction_2012))

# Рейтинг 4.9 у 2010 і 2011
books_49 = df_books[(df_books['user_rating'] == 4.9) & (df_books['year'].isin([2010, 2011]))]
print("Кількість:", len(books_49))

# Мін і макс ціна по жанрам
result = df_books.groupby('genre')['price'].agg(['min', 'max'])
print(result)
```
