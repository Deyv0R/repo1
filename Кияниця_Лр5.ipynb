```python
import plotly.express as px
import numpy as np
import pandas as pd

df = px.data.iris()
# 1. Графіки для Iris
# Дискретні кольори
fig1 = px.scatter(
    df,
    x="sepal_length",
    y="sepal_width",
    color="species",
    title="Iris Dataset (Дискретні кольори)"
)

fig1.update_layout(title_x=0.5)
fig1.show()

# Неперервні кольори
fig2 = px.scatter(
    df,
    x="sepal_length",
    y="petal_length",
    color="petal_length",
    title="Iris Dataset (Неперервні кольори)",
    color_continuous_scale="Viridis"
)

fig2.update_layout(title_x=0.5)
fig2.show()

#Різні маркери
fig3 = px.scatter(
    df,
    x="sepal_width",
    y="petal_length",
    color="species",
    symbol="species",
    title="Iris Dataset (Різні маркери)"
)

fig3.update_layout(title_x=0.5)
fig3.show()

# 2. Лінія тренду (tips)
tips = px.data.tips()

fig4 = px.scatter(
    tips,
    x="total_bill",
    y="tip",
    trendline="ols",
    title="Лінія тренду (Tips Dataset)"
)

fig4.update_layout(title_x=0.5)
fig4.show()

# 3. Gapminder (дохід vs тривалість життя)
gap = px.data.gapminder()

fig5 = px.scatter(
    gap.query("year == 2007"),
    x="gdpPercap",
    y="lifeExp",
    size="pop",
    color="continent",
    log_x=True,
    title="Доход vs Тривалість життя (2007)"
)

fig5.update_layout(title_x=0.5)
fig5.show()

# 4. Динаміка пасажирів (Японія)
# У gapminder немає пасажирів, тому беремо приклад з населенням
japan = gap[gap["country"] == "Japan"]

fig6 = px.line(
    japan,
    x="year",
    y="pop",
    title="Динаміка населення Японії"
)

fig6.update_layout(title_x=0.5)
fig6.show()

# 5. Box Plot (довжина пелюстки)
fig7 = px.box(
    df,
    x="species",
    y="petal_length",
    color="species",
    title="Розподіл довжини пелюстки"
)

fig7.update_layout(title_x=0.5)
fig7.show()

# 6. Тригонометричні функції
x = np.linspace(0, 10, 100)

#sin (x)
fig8 = px.line(
    x=x,
    y=np.sin(x),
    title="y = sin(x)"
)

fig8.update_layout(title_x=0.5)
fig8.show()
#cos (x)
fig9 = px.line(
    x=x,
    y=np.cos(x),
    title="y = cos(x)"
)

fig9.update_layout(title_x=0.5)
fig9.show()
#tan (x)
fig10 = px.line(
    x=x,
    y=np.tan(x),
    title="y = tan(x)"
)

fig10.update_layout(title_x=0.5, yaxis_range=[-10, 10])
fig10.show()

```
